```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import plotly.graph_objects as go
```

```python
df = pd.read_csv('food.csv')
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
df1 = df[['Category','Data.Vitamins.Vitamin A - IU', 'Data.Vitamins.Vitamin A - RAE',
       'Data.Vitamins.Vitamin B12', 'Data.Vitamins.Vitamin B6',
       'Data.Vitamins.Vitamin C', 'Data.Vitamins.Vitamin E',
       'Data.Vitamins.Vitamin K']]
df1.head()
```

```python
g = sns.pairplot(df1, height=2.5);
```

```python
f = sns.PairGrid(df1)
f.map_upper(plt.scatter)
f.map_lower(sns.kdeplot)
f.map_diag(sns.kdeplot, lw=3, legend=False);
```
